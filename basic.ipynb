```python
## linspace
import numpy as np 
import matplotlib.pyplot as plt 
n = 1000
xmin, xmax = -2*np.pi, 2*np.pi 
x = np.linspace(xmin, xmax, n)
y = np.sin(x)**2
## use latex in pyplot 
plt.plot(x,y, label = r'$y = \sin^2 x$'.format(2))
plt.legend(loc = 'lower center')

# plot limits 
plt.xlim(xmin, xmax)
plt.ylim(0, 1)

plt.show()
```

## Polar Plots

```python
theta = np.linspace(0,2.*np.pi, 1000)
a = 1. 
r = 2 * a * (1. + np.cos(theta))
plt.polar(theta, r)
plt.show()
```

## Multiple axes

```python
import matplotlib.pyplot as plt 

years = range(2000, 2010)
divorce_rate = np.random.rand(10) + 4
margarine_consumption = np.random.rand(10)*3 +5

line1 = plt.plot(years, divorce_rate, 'b-o', label = "Divorce rate") # return list of ojects representing
plt.ylabel('Divorces per 1000 people')


plt.twinx() # get the second y label
line2 = plt.plot(years, margarine_consumption, 'r-o', label = 'Margarine consumption')
plt.ylabel('lb of Margarine (per capita)')

lines = line1 + line2 
labels = []
for line in lines:
    labels.append(line.get_label())
    plt.legend(lines, labels)

plt.show()
```

## defaultdict 
- will return default when assigned key is missing

```python
from collections import defaultdict
word_lengths = defaultdict(int)
text ="asd sdafsdf asd sdfs af f a f asdfasdf asdf sdf df dfd ggfdg "
for word in text.split():
    # if key is missing, will be inserted with 0
    word_lengths[len(word)] += 1
print(word_lengths)
```

## Generator
- declare a function that behaves like an iterable object

```python
def count(n):
    i = 0 
    while i < n:
        i += 1
        yield i 

for j in count(5):
    print(j)

# generator comprehension syntax
print(list(count(5)))
```

## Walrus Operator


```python
s = "A simple and short text"

# Problem len(s) run two times
if len(s)> 3:
    print(f"s has {len(s)} characters. The maximum is 10")


# try solve
slen = len(s)
if slen> 3:
    print(f"s has {slen} characters. The maximum is 6")

# walrus operator 
if (slen := len(s)) > 10: # := return that value
    print(f"s has {slen} characters. The maximum is 6")
```
